# **Train TransUnet**

```python
import sys
import torch
import argparse

# Additional Scripts
from t_train_octa import TrainTestPipe
from inference import SegInference
from config_octa import cfg

def main_pipeline(model_path, train_path, valid_path):
    device = 'cpu:0'
    if torch.cuda.is_available():
        device = 'cuda:0'

    # to start train
    ttp = TrainTestPipe(train_path=train_path,
                        test_path=valid_path,
                        model_path=model_path,
                        device=device,
                        cfg = cfg)

    train_losses, test_losses = ttp.train()

    return train_losses, test_losses

if __name__ == '__main__':
    # to save the best model
    model_path = r"/content/drive/MyDrive/OCTA500/Project/TransUnet/best_model.pth"

    # define paths of datasets
    train_path = r"/content/drive/MyDrive/OCTA500/Project/TransUnet/t_train.npz"
    valid_path = r"/content/drive/MyDrive/OCTA500/Project/TransUnet/t_valid.npz"

    train_losses, test_losses = main_pipeline(model_path, train_path, valid_path)
```

### Save the historical loss values

```python
import json

# Prepare data for saving
data_to_save = {
    "train_losses": train_losses,
    "test_losses": test_losses
}

# Specify your file path
file_path = "/content/drive/MyDrive/OCTA500/Project/TransUnet/results.json"

# Save data to a file
with open(file_path, 'w') as file:
    json.dump(data_to_save, file)

print(f"Data saved to {file_path}")
```

## Plot the loss while training and validation

```python
import json
import matplotlib.pyplot as plt
%matplotlib inline

# Specify the file path
file_path = "/content/drive/MyDrive/OCTA500/Project/TransUnet/results.json"

# Load data from the file
with open(file_path, 'r') as file:
    loaded_data = json.load(file)

# Retrieve the two arrays
train_losses = loaded_data["train_losses"]
test_losses = loaded_data["test_losses"]

# Convert the data to lists if needed
train_losses = list(train_losses)
test_losses = list(test_losses)

%config InlineBackend.figure_format="svg"

# Plotting train losses
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(train_losses) + 1), train_losses, label='Train Loss', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Dice Loss')
plt.title('Train Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.savefig("lossplottrain.pdf")
plt.show()

# Plotting test losses
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(test_losses) + 1), test_losses, label='Test Loss', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Dice Loss')
plt.title('Validation Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.savefig("lossplot.pdf")
plt.show()
```

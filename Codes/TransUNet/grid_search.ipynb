```python
import itertools
import torch
from train_octa import TrainTestPipe
from config_octa import cfg

def main_pipeline(train_path, test_path, model_path, device, cfg, log_file):
    ttp = TrainTestPipe(train_path=train_path,
                        test_path=test_path,
                        model_path=model_path,
                        device=device,
                        cfg=cfg,
                        log_file=log_file)
    ttp.train()

if __name__ == '__main__':
    # Define the parameter grid
    param_grid = {
    'batch_size': [8, 16, 32],
    'learning_rate': [1e-2, 1e-3, 1e-4],
    'momentum': [0.9, 0.95],
    'weight_decay': [1e-4, 1e-5]
    }

    # Define paths and device
    train_path = "/content/drive/MyDrive/OCTA500/Project/TransUnet/train.npz"
    valid_path = "/content/drive/MyDrive/OCTA500/Project/TransUnet/test.npz"
    model_path = r"/content/drive/MyDrive/OCTA500/Project/TransUnet/best_model.pth"
    device = 'cuda:0' if torch.cuda.is_available() else 'cpu:0'

    # Open a log file
    with open('grid_search_log.txt', 'w') as log_file:
        # Iterate over all combinations of parameters
        for params in itertools.product(*param_grid.values()):
            # Update cfg with current combination
            for key, value in zip(param_grid.keys(), params):
                setattr(cfg, key, value)  # Update cfg parameter

            # Print current parameters
            current_params_str = ', '.join([f'{k}: {v}' for k, v in zip(param_grid.keys(), params)])
            print(f'Training with parameters: {current_params_str}')

            # Train and evaluate the model
            main_pipeline(train_path, valid_path, model_path, device, cfg, log_file)
```

Until that part of the grid search, we can observe that weight_decay = 1e-5 preforms better than weight_decay = 1e-4. And also, momentum = 0.95 performs better than momentum = 0.9. Because of that we have limited computtaional resources, and the grid search takes lomg time, we stopped the loop and restart new search by updating parameter grid.
